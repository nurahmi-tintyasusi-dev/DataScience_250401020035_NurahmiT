```python
# Membuat list
data = [10, 20, 'Python', True, 3.14]

# Akses & slicing
print(data[0])      # → 10
print(data[-1])     # → 3.14 (dari belakang)
print(data[1:3])    # → [20, 'Python']

# Manipulasi
data.append(99)     # tambah di akhir
data.insert(1, 55)  # sisipkan di posisi 1
data.remove(20)     # hapus nilai pertama = 20
# data.sort()         # urutkan (jika tipe sama)
```

```python
import numpy as np

# Dari list Python
a1D = np.array([1, 2, 3, 4, 5])           # 1D
a2D = np.array([[1,2,3],[4,5,6]])          # 2D

# Array dengan nilai inisial
zeros = np.zeros((3, 4))        # nol semua
ones  = np.ones((2, 3))         # satu semua
eye   = np.eye(4)               # matriks identitas
full  = np.full((2, 2), 7)      # nilai konstan

# Sequence
arange = np.arange(0, 20, 2)    # 0,2,4,...18
linsp  = np.linspace(0, 1, 11)  # 11 titik merata

# Array acak (reproducible)
rng = np.random.default_rng(seed=42)
rand_int = rng.integers(0, 100, (3, 4))
rand_flt = rng.normal(loc=0, scale=1, size=(100,))
```

```python
# Operasi aritmetika vektorisasi
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])
print(a + b)       # → [11 22 33 44 55]
print(a * b)       # → [10 40 90 160 250]
print(a ** 2)      # → [1 4 9 16 25]

# Universal functions (ufuncs)
np.sqrt(a)         # akar kuadrat tiap elemen
np.exp(a)          # eksponensial (e^x)
np.log(b)          # logaritma natural

# Fungsi statistik
data = rng.normal(0, 1, 1000)
print(f'Mean: {np.mean(data):.4f}')
print(f'Std:  {np.std(data):.4f}')
print(f'Min:  {np.min(data):.4f}')
print(f'Max:  {np.max(data):.4f}')
print(f'Q1:   {np.percentile(data,25):.4f}')
print(f'Q3:   {np.percentile(data,75):.4f}')

```

```python
import pandas as pd

# Series dari list — index default 0,1,2...
s1 = pd.Series([85, 92, 78, 95, 88])

# Series dengan index kustom
nilai = pd.Series(
    [85, 92, 78, 95],
    index=['Alice','Bob','Caca','Dina']
)
print(nilai['Bob'])     # → 92
print(nilai[nilai>80])  # filter
print(nilai.mean())     # → 87.5
print(nilai.describe()) # statistik lengkap

```

```python
# Membuat DataFrame dari dictionary
data = {
    'nama':  ['Ali','Budi','Caca','Dina'],
    'usia':  [22, 25, 21, 23],
    'kota':  ['Jakarta','Bandung','Surabaya','Medan'],
    'ipk':   [3.8, 3.5, 3.9, 3.7]
}
df = pd.DataFrame(data)

# Informasi dasar
print(df.shape)    # → (4, 4)
print(df.dtypes)   # tipe data per kolom
print(df.columns)  # nama-nama kolom
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Gunakan sintaks read_csv langsung tanpa pembungkus DataFrame ganda
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Menampilkan informasi dimensi dan beberapa baris data untuk memastikan sukses
print(f"Data Berhasil Dimuat! Ukuran: {df.shape}\n")
df.head()
```

```python
# 1. Menampilkan informasi struktur dataframe (tipe data dan baris non-null)
print("--- Informasi Struktur Data ---")
df.info()

# 2. Menampilkan dimensi/ukuran dataset (baris, kolom)
print(f"\nDimensi Dataset (Baris, Kolom): {df.shape}")

# 3. Menghitung jumlah nilai kosong (missing values) di setiap kolom
print("\n--- Jumlah Missing Values Per Kolom ---")
print(df.isnull().sum())
```

```python
# Menampilkan statistik deskriptif (mean, min, max, kuartil) untuk data numerik
print("--- Ringkasan Statistik Deskriptif ---")
print(df.describe())
```

```python
# 1. Menghitung rata-rata tingkat keselamatan (Survived) berdasarkan Jenis Kelamin (Sex)
print("--- Rata-rata Keselamatan Berdasarkan Jenis Kelamin ---")
# Catatan: Kolom Survived berisi 1 (Selamat) dan 0 (Tidak Selamat)
print(df.groupby("Sex")["Survived"].mean())

# 2. Menghitung jumlah penumpang berdasarkan Kelas Tiket (Pclass)
print("\n--- Jumlah Penumpang Per Kelas Tiket (Pclass) ---")
print(df["Pclass"].value_counts())
```

```python
# 1. Mengatur tema visualisasi
sns.set_theme(style="whitegrid")

# 2. Membuat Figure dengan ukuran 12x5 inci (berisi 2 sub-plot berdampingan)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 3. Grafik Kiri: Tingkat Keselamatan Berdasarkan Jenis Kelamin
# Ditambahkan hue="Sex" dan legend=False untuk menghilangkan FutureWarning
sns.barplot(x="Sex", y="Survived", data=df, hue="Sex", errorbar=None, palette="muted", legend=False, ax=axes[0])
axes[0].set_title("Persentase Keselamatan Berdasarkan Jenis Kelamin")
axes[0].set_ylabel("Rata-rata Selamat (0 sampai 1)")
axes[0].set_xlabel("Jenis Kelamin")

# 4. Grafik Kanan: Jumlah Penumpang Berdasarkan Kelas Tiket (Pclass)
sns.countplot(x="Pclass", data=df, hue="Pclass", palette="deep", legend=False, ax=axes[1])
axes[1].set_title("Jumlah Penumpang Per Kelas Tiket (Pclass)")
axes[1].set_ylabel("Jumlah Penumpang")
axes[1].set_xlabel("Kelas Tiket (1 = Atas, 2 = Menengah, 3 = Bawah)")

# 5. Menampilkan grafik secara rapi
plt.tight_layout()
plt.show()
```
