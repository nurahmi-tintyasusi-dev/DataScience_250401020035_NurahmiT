# PERTEMUAN 3 - Housing Dirty Dataset

```python
# I port Library untuk data cleaning

import pandas as pd
import numpy as np
dataset_dirty = pd.read_csv('./housing_dirty.csv')
df = dataset_dirty

# Load Dataset housing
print(df.info())
```

```python
#Memuat data awal
print(df.describe)
```

```python
#Memuat data kosong

print(df.isnull().sum())
print('shape awal:',df.shape)
```

Langkah 1: Menghapus data duplikat

```python
df.drop_duplicates(inplace=True)
print('setelah hapus duplikat:',df.shape)
```

Langkah 2: Normalisasi teks pada kolom kategorikal

```python
df['harga_juta']=df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar']=df['kamar'].fillna(df['kamar'].mode()[0])
df['luas_m2']=df['luas_m2'].fillna(df['luas_m2'].median())
df['kota']=df['kota'].str.strip().str.title()
df['kondisi']=df['kondisi'].str.strip().str.lower()
```

Langkah 3: Menangani missing values

```python
df.fillna({'luas_m2': df['luas_m2'].median()}, inplace=True)
df.fillna({'harga_juta': df['harga_juta'].median()}, inplace=True)
df.fillna({'kamar': df['kamar'].mode()[0]}, inplace=True)
```

Langkah 4: Menangani outlier dengan metode IQR

```python
for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:
  Q1, Q3 = df[col].quantile([0.25, 0.75])
  IQR = Q3 - Q1
  df[col]= df[col].clip(lower=Q1 - 1.5 * IQR, upper=Q3 + 1.5 * IQR)
```

Langkah 5: Validasi dan ekspor data bersih

```python
assert df.isnull().sum().sum()== 0, 'Masih ada missing!'
assert df.duplicated().sum()== 0, 'Masih ada duplikat!!'
print ('shape akhir:',df.shape)
df.to_csv ('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!!')
```

Kesimpulan


*   **Apa yang dipelajari :** Proses data cleaning : menghapus duplikat, Normalisasi teks pada kolom kategorikal, menangani missing values, Menangani outlier dengan metode IQR
*   **Temuan utama :** Pembersihan data secara terstruktur sukses membuang inkonsistensi format, nilai tidak valid, atau skala yang tidak seragam. Menyimpan data hasil akhir yang telah bersih ke dalam file ekspor baru (seperti housing_clean.csv)
*   **Keterbatasan/Pertanyaan:** Mengubah Distribusi Data: Memotong nilai ekstrem secara paksa (menggunakan IQR) dapat mengubah bentuk asli distribusi data secara drastis, terutama jika data awalnya sangat condong, Kehilangan Informasi Berharga:*outlier* tidak selalu merupakan kesalahan input data. Menghapusnya berisiko menghilangkan informasi tentang kejadian ekstrem nyata yang penting bagi bisnis atau sistem.
Apakah Logis Secara Realitas? Mengisi data kosong (missing values) menggunakan nilai tebakan statistik (mean, median, atau modus) berisiko menciptakan data palsu yang tidak masuk akal secara realitas bisnis.
