```python
# Import library untuk data cleaning

import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize

# Load dataset housing
df_housing = pd.read_csv('housing_dirty.csv')
print('Shape awal:', df_housing.shape)
print(df_housing.info())
print(df_housing.describe())
print(df_housing.isnull().sum())

# Langkah 0: Memuat dan eksplorasi data awal
df_housing.drop_duplicates(inplace=True)
print(f"Ukuran setelah hapus duplikat: {df_housing.shape}")

df_housing['kota'] = df_housing['kota'].str.strip().str.title()
df_housing['kondisi'] = df_housing['kondisi'].str.strip().str.lower()


# Imputasi missing values
df_housing['luas_m2'].fillna(df_housing['luas_m2'].median(), inplace=True)
df_housing['harga_juta'].fillna(df_housing['harga_juta'].median(), inplace=True)
df_housing['kamar'].fillna(df_housing['kamar'].mode()[0], inplace=True)

# Clipping outlier menggunakan IQR
for kolom in ['harga_juta', 'luas_m2', 'tahun_bangun']:
    q1 = df_housing[kolom].quantile(0.25)
    q3 = df_housing[kolom].quantile(0.75)
    iqr = q3 - q1
    batas_bawah = q1 - 1.5 * iqr
    batas_atas = q3 + 1.5 * iqr
    df_housing[kolom] = df_housing[kolom].clip(batas_bawah, batas_atas)
    print(f"{kolom}: Q1={q1:.2f}, Q3={q3:.2f}, IQR={iqr:.2f}")


# Validasi data
total_missing = df_housing.isnull().sum().sum()
total_duplikat = df_housing.duplicated().sum()

if total_missing == 0 and total_duplikat == 0:
    print(f"Ukuran akhir: {df_housing.shape}")
    df_housing.to_csv('housing_clean.csv', index=False)
    print("Data bersih berhasil disimpan!")
else:
    print(f"Masih ada masalah: Missing={total_missing}, Duplikat={total_duplikat}")


```
